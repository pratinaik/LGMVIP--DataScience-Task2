**Image to Pencil Sketch with Python - TASK 2**

To programmatically transform a color image into a black and white pencil sketch with enhanced details, we first read the image in RGB format. Convert the image to grayscale, removing color information. Then create an inverted grayscale image by reversing the intensity values. Blur the inverted image to enhance artistic appearance. Blend the grayscale and inverted blurry images by dividing corresponding pixel values. The resulting image will resemble a pencil sketch.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
image_path = "/content/MahadevJi.png"
image = cv2.imread(image_path)
resized_image = cv2.resize(image, (500, 500))
cv2_imshow(resized_image)
```

```python
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
grayImage
```

```python
plt.imshow(grayImage)
```

Inverting the above grayscale image

```python
invertedImage = 255 - grayImage
plt.imshow(cv2.cvtColor(invertedImage, cv2.COLOR_BGR2RGB))
```

Blur the image by using the Gaussian Function in OpenCV

```python
blurred = cv2.GaussianBlur(invertedImage, (21,21), 0)
plt.imshow(blurred)
```

Inverting the above blurred image to get the pencil sketch

```python
invertedBlur = 255 - blurred
pencil_sketch = cv2.divide(grayImage, invertedBlur, scale=256.0)
plt.imshow(pencil_sketch)
```

Final Sketch

```python
invertedBlur = 255 - blurred
pencil_sketch = cv2.divide(grayImage, invertedBlur, scale=256.0)
plt.imshow(cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
